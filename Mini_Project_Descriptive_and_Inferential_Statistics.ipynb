ğŸ“Š Foundational Knowledge
**1. Key Statistical Definitions**

Mean (Average): Sum of all values Ã· number of values.

Median: Middle value when data is sorted.

Mode: Most frequently occurring value.

Variance: Measure of how spread out the data is from the mean.

Standard Deviation (Ïƒ or s): Square root of variance, shows spread in the same units as data.

Probability: Likelihood of an event occurring (between 0 and 1).

Correlation: Measures the relationship between two variables (ranges from â€“1 to +1).

**2. Data Types**

Quantitative (Numerical):

Continuous: Can take any value within a range (e.g., height, temperature).

Discrete: Countable values (e.g., number of students, goals scored).

Qualitative (Categorical):

Nominal: Categories without order (e.g., colors, gender).

Ordinal: Categories with a meaningful order (e.g., rankings, satisfaction levels).

**3. Population vs Sample**

*Population:* The entire group of individuals or observations of interest.

Example: All the students in a university.

*Sample:* A smaller subset of the population, used to make inferences.

Example: 200 randomly chosen students from the university.

Key Idea:

Studying the whole population is often impractical.

We use a sample to estimate population characteristics.

These estimates are subject to sampling error (differences between sample results and true population values).

a) Nominal

Definition: Data is categorized into distinct groups without any order or ranking.

b) Ordinal

Definition: Data is categorized with a meaningful order, but the differences between values are not uniform or measurable.

Example: Customer satisfaction ratings (Poor, Fair, Good, Excellent).

c) Interval

Definition: Data has ordered categories with equal intervals between values, but no true zero point.

Example: Temperature in Celsius (0Â°C doesnâ€™t mean â€œno temperatureâ€).

d) Ratio

Definition: Data has ordered categories, equal intervals, and a true zero point (zero means absence of the quantity). Ratios are meaningful.

Example: Weight (0 kg means no weight, and 20 kg is twice 10 kg).
