📊 Foundational Knowledge
**1. Key Statistical Definitions**

Mean (Average): Sum of all values ÷ number of values.

Median: Middle value when data is sorted.

Mode: Most frequently occurring value.

Variance: Measure of how spread out the data is from the mean.

Standard Deviation (σ or s): Square root of variance, shows spread in the same units as data.

Probability: Likelihood of an event occurring (between 0 and 1).

Correlation: Measures the relationship between two variables (ranges from –1 to +1).

**2. Data Types**

Quantitative (Numerical):

Continuous: Can take any value within a range (e.g., height, temperature).

Discrete: Countable values (e.g., number of students, goals scored).

Qualitative (Categorical):

Nominal: Categories without order (e.g., colors, gender).

Ordinal: Categories with a meaningful order (e.g., rankings, satisfaction levels).

**3. Population vs Sample**

*Population:* The entire group of individuals or observations of interest.

Example: All the students in a university.

*Sample:* A smaller subset of the population, used to make inferences.

Example: 200 randomly chosen students from the university.

Key Idea:

Studying the whole population is often impractical.

We use a sample to estimate population characteristics.

These estimates are subject to sampling error (differences between sample results and true population values).

# **Q4. Levels of Measurement**

Objective: Understand classification of data types.

a) Nominal

Definition: Data is categorized into distinct groups without any order or ranking.

Example: Blood groups (A, B, AB, O).

b) Ordinal

Definition: Data is categorized with a meaningful order, but the differences between values are not uniform or measurable.

Example: Customer satisfaction ratings (Poor, Fair, Good, Excellent).

c) Interval

Definition: Data has ordered categories with equal intervals between values, but no true zero point.

Example: Temperature in Celsius (0°C doesn’t mean “no temperature”).

d) Ratio

Definition: Data has ordered categories, equal intervals, and a true zero point (zero means absence of the quantity). Ratios are meaningful.

Example: Weight (0 kg means no weight, and 20 kg is twice 10 kg).

✅ Quick Check:

Nominal = Names only

Ordinal = Order matters

Interval = Order + equal spacing, but no true zero

Ratio = Order + equal spacing + true zero

# ***📘 Q8. Five-number Summary and Boxplot Concept***

Objective: Summarize distribution of data.

Five-number Summary

The five-number summary is a set of descriptive statistics that provides a quick overview of a dataset’s distribution. It consists of:

Minimum

The smallest value in the dataset.

Shows the lower bound of the data (excluding outliers).

Example: In {5, 8, 12, 20, 25}, minimum = 5.

Q1 (First Quartile)

The 25th percentile of data (median of the lower half).

25% of values lie below Q1.

Example: Q1 = 8 in dataset above.

Median (Q2)

The middle value of the dataset (50th percentile).

Divides data into two equal halves.

Example: Median = 12.

Q3 (Third Quartile)

The 75th percentile of data (median of the upper half).

75% of values lie below Q3.

Example: Q3 = 20.

Maximum

The largest value in the dataset.

Shows the upper bound of the data (excluding outliers).

Example: Maximum = 25.

Boxplot (or Whisker Plot) and Outliers

A boxplot visually represents the five-number summary:

Box: Extends from Q1 to Q3.

Line inside box: Median.

Whiskers: Extend to minimum and maximum (within range).

Outlier detection:

Outliers are usually defined as data points that fall below Q1 – 1.5×IQR or above Q3 + 1.5×IQR.

These appear as individual dots or stars beyond whiskers in a boxplot.

Example:
If Q1 = 10, Q3 = 20 → IQR = 10.

Lower fence = 10 – 1.5×10 = –5

Upper fence = 20 + 1.5×10 = 35

Any value < –5 or > 35 is considered an outlier.

✅ Key Takeaway:

Five-number summary = quick numeric overview.

Boxplot = graphical summary + easy way to spot skewness and outliers.
